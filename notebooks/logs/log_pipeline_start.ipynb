```python
# Databricks notebook source
# Receber parâmetros do ADF
dbutils.widgets.text("execution_id", "")
dbutils.widgets.text("pipeline_name", "")
dbutils.widgets.text("executed_by", "SYSTEM")  # ← ADICIONAR

execution_id = dbutils.widgets.get("execution_id")
pipeline_name = dbutils.widgets.get("pipeline_name")
executed_by = dbutils.widgets.get("executed_by")  # ← ADICIONAR

# Logar início
from datetime import datetime

spark.sql("USE CATALOG hive_metastore")
spark.sql("USE healthcare_bronze")

spark.sql(f"""
INSERT INTO pipeline_execution_log VALUES (
    '{execution_id}',
    '{pipeline_name}',
    current_timestamp(),
    NULL,
    'RUNNING',
    0,
    NULL,
    current_date(),
    '{executed_by}'
)
""")

print(f"Pipeline {pipeline_name} iniciada - Execution ID: {execution_id}")
dbutils.notebook.exit("SUCCESS")
```
